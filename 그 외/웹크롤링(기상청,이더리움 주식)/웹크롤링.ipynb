```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as offline
import plotly.graph_objs as go
data = requests.get("https://crix-api-endpoint.upbit.com/v1/crix/candles/minutes/30?code=CRIX.UPBIT.KRW-ETH&count=210")
json=data.json()
df=pd.DataFrame(json)
item_name="이더리움"
df

trace = go.Scatter(x=df["candleDateTime"], y=df["tradePrice"], name=item_name)
# 위에 데이터 정보를 data라는 객체의 리스트로 담아준다.
data = [trace]
# 레이아웃 잡기
layout = dict(title='{}의 종가(close) Time Series'.format(item_name),  # 타이틀 생성.
              xaxis=dict(
                  rangeselector=dict(
                      buttons=list([  
                          dict(
                              count=1,  # 1개씩 센다. 여기서는 step='month'이기 때문에 1달이 된다.
                              label='1d',  # 라벨 이름. 그래프에 1m이라는 버튼을 만든다.
                              step='day',  # 한 달을 기준으로 잡아서 count를 센다.
                              stepmode='backward'),  # 가장 최근 데이터부터 센다. todate는 가장 오래된 데이터부터 센다.
                          dict(count=3, 
                               label='3d', 
                               step='day', 
                               stepmode='backward'),

                          dict(
                              step='all')
                      ]) ),  # 전체 데이터를 출력한다. step='all'은 label을 설정할 수 없다.
                  rangeslider=dict(),
                  type='date'
                    )
              )
fig = go.Figure(data=data, layout=layout)
# 플롯을 출력한다.
offline.iplot(fig)
offline.plot(fig, filename = 'C:/Users/Hyeonmo/Desktop/Face/temp_eth.html')
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
df=pd.DataFrame(columns=['Region','Rate'])
html=urlopen("http://www.weather.go.kr/weather/lifenindustry/li_asset/popup/imgdata_popup.jsp?CODE=D01&point=0")
bsObject=BeautifulSoup(html,"html.parser")
table=bsObject.body.find_all("table",{"class":"viewDataT"})
index=table[0]
trs=index.find_all('tr')
for i in range(2,len(trs)):
    region=trs[i]
    ths=region.find_all('th',{"scope":"row"})
    tds=region.find_all('td')
    for j in range(0,len(ths)):
        df2=pd.DataFrame({'Region':[ths[j].text], \
                      'Rate':[tds[j].text]})
        df=df.append(df2,ignore_index=True)

df
```
